```python
from augmentlib.methods.aga import AGAAugmentor
from PIL import Image
```

```python
aga = AGAAugmentor(
    class_name="flower",       # обязательно
    device="cuda",
    load_models=True           # модели загрузятся сразу
)
```

```python
img = Image.open("image_03479.jpg").convert("RGB")
```

```python
aug_img = aga.augment(img)
```

```python
if aug_img is not None:
    aug_img.save("augmented_flower.jpg")
```

```python
from augmentlib.methods.aga import AGAAugmentor
from augmentlib.pipelines.offline import OfflineAugmentedDatasetGenerator
```

```python
from local_datasets.oxford_flower102 import OxfordFlowers102
```

```python
flowers = OxfordFlowers102(root='./data', transform=None, download=True)
```

```python
print(f"Train: {len(flowers.train_dataset)}")
print(f"Val:   {len(flowers.val_dataset)}")
print(f"Test:  {len(flowers.test_dataset)}")
```

```python
from augmentlib.pipelines.offline import OfflineAugmentedDatasetGenerator

generator = OfflineAugmentedDatasetGenerator(
    aga,
    save_dir="./aga",
    num_aug=1
)
```

```python
generator.generate(flowers.train_dataset)
```
